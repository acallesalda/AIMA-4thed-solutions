```python
# Exercise 2.11

import numpy as np
import random

# Vacuum world

performances = np.zeros(100)
for j in range(1,100): # 100 different experiments
    # initialize locations randomly
    locations = [random.choice([0, 1]), random.choice([0, 1])]
    # 1 is clean, 0 dirty
    for i in range(1,10): # Lets say the algorithm will run for 10 steps
        print('No agent yet, i.e. this agent does nothing')
    print(locations)
    performances[j] = sum(locations) # amount of clean squares after 10 steps
```

```python
# Mean performance of the agent (in this case an agent that does nothing)
np.mean(performances)
```

```python
# Exercise 2.12

#All potential configurations of squares
configs = [[0, 0], [0, 1], [1, 0], [1, 1]]
#Potential starting point
start_robot = [0, 1]
performances = []
for config in configs:
    for start in start_robot:
        s = start
        for i in range(1,10):
            if config[s] == 0:
                config[s] = 1
            if s == 0:
                s = 1
            else:
                s = 0
        performances.append(np.sum(config))
```

```python
# Number of clean spaces after all runs. 
np.mean(performances)
```

```python
# Exercise 2.13

#All potential configurations of squares
configs = [[0, 0], [0, 1], [1, 0], [1, 1]]
#Potential starting point
start_robot = [0, 1]
performances = []
states = [0, 0]
for config in configs:
    for start in start_robot:
        s = start
        i = 0
        # if the 2 states are clean, the agent will stop
        # moving, incurring in more cost. 
        while sum(states) < 2 and i <= 10:
            if config[s] == 0:
                config[s] = 1
            states[s] = 1
            if s == 0:
                s = 1
            else:
                s = 0
            i = i + 1
        performances.append(np.sum(config) - i)
```

```python
np.mean(performances)
```

```python
# Exercise 2.14

# Implement random envioroment 
width = np.random.choice([x for x in range(1,10)])
heigth = np.random.choice([x for x in range(1,10)])
envir = np.random.choice([0,1], size = (heigth, width))

# Place the agent randomly.

y = np.random.choice([x for x in range(0, heigth-1)])
x = np.random.choice([x for x in range(0, width-1)])

# Implement a randomized agent function, take into account
# illegal moves. 

envir_start = envir
```

```python
# time steps
T = 1000
for t in range(1,T):
    # Check current coordinate, clean if dirty
    if envir[x, y] == 0:
        envir[x, y] = 1
    # Check for legal moves.
    legal = []
    # Convention: 1 = up, 2 = down, 3 = left, 4 = right
    #print(x_coord)
    #print(y_coord)
    if y != 0:
        legal.append(1)
    if y != heigth - 1:
        legal.append(2)
    if x != 0:
        legal.append(3)
    if x != width - 1:
        legal.append(4)
    #print(legal)
    
    # Pick a move randomly.
    move = np.random.choice(legal)
    
    # Change coordinate based on move
    if move == 1:
        y = y + 1
    if move == 2:
        y = y - 1
    if move == 3:
        x = x - 1
    if move == 4:
        x = x + 1
        
# TENGO QUE DEBUGGEAR ESTO, VER MOVES ILEGALES ESTA MAL
```
